```python
import matplotlib.pyplot as plt
import numpy as np

# Task labels and data for each model
tasks = ['ASR (wo LM)', 'KS', 'IC', 'SingerID', 'VocalID']
hubert_scores = [6.42, 96.3, 98.34, 81.49, 65.35]
mert_scores = [None, None, 59.21, 85.69, 72.54]
average_scores = [None, 57.0593963, 16.76772982, 52.17082, 47.71929681]

# Convert scores to numpy arrays for easier manipulation
hubert_scores = np.array(hubert_scores, dtype=np.float64)
mert_scores = np.array(mert_scores, dtype=np.float64)
average_scores = np.array(average_scores, dtype=np.float64)

# Define bar width and positions
bar_width = 0.25
r1 = np.arange(len(tasks))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Plotting
plt.figure(figsize=(10, 6))

plt.bar(r1, hubert_scores, color='blue', width=bar_width, edgecolor='grey', label='HuBERT')
plt.bar(r2, mert_scores, color='orange', width=bar_width, edgecolor='grey', label='MERT')
plt.bar(r3, average_scores, color='green', width=bar_width, edgecolor='grey', label='HuBERT-MERT Average')

# Adding labels
plt.xlabel('Tasks', fontweight='bold')
plt.ylabel('Performance', fontweight='bold')
plt.title('Performance Comparison of Models on Different Tasks', fontweight='bold')

# X-axis task labels
plt.xticks([r + bar_width for r in range(len(tasks))], tasks)

# Adding legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Updated data for the new models and tasks

# Model names and performance scores for each task
models = [
    'distilhubert-init-hubert-libri-960',
    'distill_only_mert-init-weight-from-hubert_base-models-simple-avg-pool-for-teacher-train-libri-960',
    'distill_mert_init_mert_music4all_avgpool',
    'mert-distilhubert-all-from-ls960-init-from-hubert-averaged',
    'mert-from-music4all-and-mert-init-distilhubert-from-hubert-and-ls960'
]
tasks = ['ASR (wo LM)', 'KS', 'IC', 'SingerID', 'VocalID']

# Corresponding scores for each model in each task
scores = [
    [13.52328058, 96.07, 94.71, 71.39, None],
    [23.214014, 89.16, 59.58, 78.36298943, None],
    [None, 86.17331982, 26.73345506, 84.555161, 68.15789342],
    [20.477, 92.83, 77.37938, 72.38433957, 64.12280798],
    [None, 56.92956832, 15.66042751, None, 43.42105389]
]

# Converting scores to a numpy array for ease of manipulation
scores = np.array(scores, dtype=np.float64)

# Define bar width and positions
bar_width = 0.15
r = [np.arange(len(tasks)) + i * bar_width for i in range(len(models))]

# Plotting
plt.figure(figsize=(15, 12))
colors = ['blue', 'orange', 'green', 'purple', 'red']

# Create a bar for each model in each task position
for i, (model, color) in enumerate(zip(models, colors)):
    plt.bar(r[i], scores[i], color=color, width=bar_width, edgecolor='grey', label=model)

# Adding labels
plt.xlabel('Tasks', fontweight='bold')
plt.ylabel('Performance', fontweight='bold')
plt.title('Performance Comparison of Distilled Models on Different Tasks', fontweight='bold')

# X-axis task labels
plt.xticks([x + (bar_width * 2) for x in range(len(tasks))], tasks)

# Adding legend
plt.legend(loc='upper left', bbox_to_anchor=(1,1))

# Display the plot
plt.tight_layout()
plt.show()
```
